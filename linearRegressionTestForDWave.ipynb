```python
poly = sym.sympify('0')

# poly += getFloatValue(0)
for i in range(VARIABLE_NUM):
    # poly += geti(i)
    poly += 2 ** 38 * ((VARIABLE_Y[i] - (VARIABLE_X[i] * getFloatValue(0) + getFloatValue(26))) ** 2)

poly = poly.expand()
new_poly = str(poly)
new_poly = re.sub('\*\*\d*', '', new_poly)
poly = sym.sympify(new_poly)
poly = poly.simplify()

print(poly)
```
