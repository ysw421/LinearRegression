```python
import random
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
VARIABLE_NUM = 1000000   # 데이터 개수
A = 0.5                    # 표준 기울기
B = 20                     # 표준 절편
RANDOM = 20              # 랜덤 오차 범위
RANGE_X_MIN = 0           # x 범위 최소값
RANGE_X_MAX = 100         # x 범위 최대값
```

```python
result_x = []
result_y = []
for i in range(VARIABLE_NUM):
    x = random.randrange(RANGE_X_MIN, RANGE_X_MAX)
    y = A * x + B + random.randrange(-RANDOM, RANDOM)
    result_x.append(x)
    result_y.append(y)
# print(result_x)
# print(result_y)
t = np.arange(RANGE_X_MIN, RANGE_X_MAX, 0.2)

plt.scatter(result_x, result_y)
plt.plot(t, A*t+B, 'r--')
plt.show()

with open("dot_data_x.pickle","wb") as fw:
    pickle.dump(result_x, fw)
    
with open("dot_data_y.pickle","wb") as fw:
    pickle.dump(result_y, fw)
```
